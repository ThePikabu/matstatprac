```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import TheilSenRegressor
from sklearn import metrics
```

```python
# 1.1
# аппроксимируем облако точек прямой
M = [10, 20, 50]
a = 2
b = 4
sigm = [1, a / 9]

for sigma in sigm:
    for m in M:

        # генерируем массив X, где Xk = k/m
        X = np.arange(0, 1, 1 / m)

        eps = np.random.normal(0, sigma, m)
        # Вычислим значения Yi = aXi + b + εi
        Y = a * X + b + eps

        # найдите оценки для коэффициентов прямой методом МНК
        model = LinearRegression().fit(X.reshape(-1, 1), Y)
        y_pred = model.predict(X.reshape(-1, 1))

        print('Для sigma = ', sigma,  ', m = ', m)

        # облако точек
        plt.scatter(X, Y, label = 'облако точек')
        # исходная прямая
        plt.plot(X, a * X + b,  label = 'исходная прямая')
        # полученная прямая
        plt.plot(X, y_pred,  label = 'полученная прямая')
        plt.title('sigma = ' + str(sigma) + ', m = ' + str(m))
        plt.legend()
        plt.show()

        print("Настоящий коэффициент a = ", a, ", полученный a = ", float(model.coef_))
        print("Настоящий коэффициент b = ", b, ", полученный b = ", model.intercept_)



        #print("Model score: ", model.score(X.reshape(-1, 1), Y))
        print('Mean Squared Error:', metrics.mean_squared_error(Y, y_pred))
```

```python
# 2.1 заменим одну точку

sigma = 1
M = [10, 20, 50]
# заменим одну точку
for m in M:
    K = [m, m / 2]
    for k in K:

        X = np.arange(0, 1, 1 / m)
        eps = np.random.normal(0, sigma, m)
        Y = a * X + b + eps

        model = LinearRegression().fit(X.reshape(-1, 1), Y)
        y_pred1 = model.predict(X.reshape(-1, 1))

        Y[round(k - 1)] = a * X[round(k - 1)] + b + 5 * sigma

        model = LinearRegression().fit(X.reshape(-1,1), Y)
        y_pred2 = model.predict(X.reshape(-1,1))

        # облако точек
        plt.scatter(X, Y, label='облако точек')
        # исходная прямая
        plt.plot(X, y_pred1, label='прямая МНК, построенная по исходным данным')
        # полученная прямая
        plt.plot(X, y_pred2, label='прямая МНК, полученная заменой одной точки')
        plt.title('Замена одной точки (МНК) при m = ' + str(m) + ', k = ' + str(k))
        plt.legend()
        plt.show()
```

```python
# 2.1 заменим две точки

sigma = 1
M = [10, 20, 50]
for m in M:
    l = m
    K = [m - 1, 0]
    for k in K:

        X = np.arange(0, 1, 1 / m)
        eps = np.random.normal(0, sigma, m)
        Y = a * X + b + eps

        model = LinearRegression().fit(X.reshape(-1, 1), Y)
        y_pred1 = model.predict(X.reshape(-1, 1))

        Y[round(k - 1)] = a * X[round(k - 1)] + b + 5 * sigma
        Y[round(l - 1)] = a * X[round(l - 1)] + b - 5 * sigma


        model = LinearRegression().fit(X.reshape(-1,1), Y)
        y_pred2 = model.predict(X.reshape(-1,1))

        # облако точек
        plt.scatter(X, Y, label='облако точек')
        # исходная прямая
        plt.plot(X, y_pred1, label='прямая МНК, построенная по исходным данным')
        # полученная прямая
        plt.plot(X, y_pred2, label='прямая полученная заменой двух точек МНК')
        plt.title('Замена двух точек (МНК) при m = ' + str(m) + ', k = ' + str(k) + ', l = ' + str(l))
        plt.legend()
        plt.show()
```

```python
# 3.1
# метод МНК для подгонки многочлена
sigma = 1

l = 4
D = [l, l + 1, l + 3]
# 1 var
# y = x**4 − 3*x**2 + x + 1
a, b, c, e, f = 1, 0, -3, 1, 1
m = 50

for d in D:
        X = np.arange(0, 1, 1 / 100)
        x = X
        eps = np.random.normal(0, sigma, 100)

        X2 = [i**2 for i in X]
        X3 = [i**3 for i in X]
        X4 = [i**4 for i in X]
        aa = [[1], [0], [-3], [1], [1]]

        ons = np.ones(len(X))

        X = [X4, X3, X2, X, ons]
        X = np.transpose(X)
        # X = [np.transpose(X4), np.transpose(X3), np.transpose(X2), np.transpose(X), [1]*len(X)]
        #print(X)
        eps = eps.reshape(-1,1)

        Y = X.dot(aa) + eps

        #Y = a * X**4 + b * X**3 + c * X**2 + e * X**1 + f + eps

        # найдите оценки для коэффициентов прямой методом МНК
        model = LinearRegression().fit(X, Y)
        y_pred = model.predict(X)

        # облако точек
        plt.scatter(x, Y, label = 'облако точек')
        # исходная прямая
        plt.plot(x, X.dot(aa),  label = 'исходная кривая')
        # полученная прямая
        plt.plot(x, y_pred,  label = 'полученная кривая')
        plt.title(' d = ' + str(d))
        plt.legend()
        plt.show()
        print( model.coef_)
        print( model.intercept_)
```
