```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

# Задача 2

```python
x1 = np.random.normal(0,1,(1000, 300))
x2 = np.random.standard_(-1, 1, (1000, 0))
x = np.concatenate((x1, x2), axis = 1)
x.shape
```

```python
mean = np.expand_dims(np.mean(x, axis = 1), axis=1)
std = np.expand_dims(np.std(x, axis=1), axis = 1)
```

```python
ans = []
for xx in x:
    ans.append(stats.anderson(xx))
ans = np.array(ans)
ans.shape
```

Если статистика для конкретной выборки больше соотвествующего критического значения, то нулевая гипотеза может быть отвергнута

```python
statistics = np.expand_dims(ans[:,0], axis = 1)
for i in range(5):
    print(ans[0][2][i],"%: ",np.count_nonzero(statistics>ans[:,1][0][i]))
```

```python
ans[1][2]
```

```python
test = np.random.normal(10, 15, (10000, 300))
test_stats = []
for tt in test:
    test_stats.append(stats.anderson(tt))
test_stats = np.array(test_stats)[:,0]
```

```python
pvalues = []
for st in statistics:
    pvalues.append(np.count_nonzero(st < test_stats))
pvalues = sorted(np.array(pvalues)/len(test))
```

```python
plt.plot(pvalues, np.arange(0, len(pvalues))/len(pvalues))
plt.plot([0,1], [0,1])
plt.show()
```

```python
plt.hist(pvalues)
plt.show()
```

```python
from skgof import ad_test
```

```python
result = []
for xx in x:
    result.append(ad_test(xx, stats.norm(np.mean(xx), np.std(xx))))
result = sorted(np.array(result)[:,1])
```

```python
plt.plot(result, np.arange(0, len(result))/len(result))
plt.plot([0,1], [0,1])
plt.show()
```

```python
plt.hist(result)
plt.show()
```
